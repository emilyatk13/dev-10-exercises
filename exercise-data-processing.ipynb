Remove rows where at least 50% of the values are missing.

```python
columns = len(songs.columns)
null_rows = []

for i in range(len(songs)):
    null_values = songs.iloc[i].isna().sum()
    if null_values >= columns/2:
        print(f'Row {i}, {null_values}')
        null_rows.append(i)
```

```python
songs = songs.drop(null_rows, axis=0).copy()
```

Let's define an outlier as a value at least 3 standard deviations away from the mean. Which columns have outliers? What are those values?

```python
#revised after viewing other notebooks
def get_outliers(column):
    three_sd = 3*(songs[column].std())
    mean = songs[column].sum()/len(songs)
    max = mean + three_sd
    min = mean - three_sd
    vals = []
    for i in range(len(songs)):
        val = songs[column].iloc[i]
        if val > max:
            vals.append(val)
        elif val < min:
            vals.append(val)
    print(column, vals)
```

```python
for column in songs.describe().columns:
    get_outliers(column)
```
